First attempt to plot datapoints in 2-D and 3-D space after clustering

```python
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

```python
vec_a = [2, 3, 4]
vec_b = [3, 1, 5]
vec_c = [7, 5, 2]
vec_d = [8, 7, 1]
vec_e = [3, 8, 6]
vec_f = [4, 9, 8]
vec_g = [9, 2, 3]
vec_h = [10, 1.8, 4]
data_points = [vec_a, vec_b, vec_c, vec_d, vec_e, vec_f, vec_g, vec_h]

# Set environment variable to mitigate memory leak issue
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
# Create KMeans object with n_init parameter
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(data_points)
clusters = kmeans.cluster_centers_

# Plotting the 3D data points and cluster centers
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(*zip(*data_points), color='blue', label='Data Points')
ax.scatter(*zip(*clusters), color='red', marker='x', s=100, label='Cluster Centers')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('KMeans Clustering in 3D')
ax.legend()
plt.show()
```
